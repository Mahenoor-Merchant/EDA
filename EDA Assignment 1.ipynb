### 2. How did you handle missing data in the wine quality data set during the feature engineering process?
### Discuss the advantages and disadvantages of different imputation techniques.

```python
wine.isnull().sum()
```

####  As there is no missing data is the wine dataset we don't need to handle the same! 

However, I am listing a few methods in which the missing data can be handled.
##### 1. Mean - We can find out the mean for the feature with missing data and fill the null datapoints with the mean.
##### 2. Median - We should Median if there are outlying values in the feature.
##### 3. Mode - In case of Categorical features,we can use Mode to fill the null columns.

#### Mean
* Advantage : The null column gets filled with the average value of the feature.
* Disadvantage : Is not a good way to fill the null datapoints in case if there are outliers present.

#### Median
* Advantage : Can be used when there are outliers in the dataset. As the Outlying values doesn't have too much affect on Median.
* Disadvantage : Can be less accurate.
    
#### Mode 
* Advantage : Is useful for filling the null values in case categorical features.
* Disadvantages : This can be useful for categorical features, but it can be less accurate than mean or median imputation for numerical features.

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np


wine_data =wine

# Separate features from the target variable, if necessary
X = wine_data.drop('quality', axis=1)
X = wine_data  # Assuming the dataset includes only features

# Standardize the data
X_std = (X - X.mean()) / X.std()

# Apply PCA
pca = PCA(n_components=X_std.shape[1])
X_pca = pca.fit_transform(X_std)

# Compute explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratio)

# Find the minimum number of components that explain 90% variance
num_components = np.argmax(cumulative_variance >= 0.90) + 1

print("Minimum number of components required to explain 90% variance:", num_components)
```
